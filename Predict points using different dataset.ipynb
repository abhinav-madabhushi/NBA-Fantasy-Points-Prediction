# Label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_pos = LabelEncoder() 
```

```python
data['Pos'] = label_encoder_pos.fit_transform(data['Pos']) 
```

```python
label_encoder_tm = LabelEncoder() 
```

```python
data['Tm'] = label_encoder_tm.fit_transform(data['Tm']) 
```

```python
label_encoder_year = LabelEncoder() 
```

```python
data['Year'] = label_encoder_tm.fit_transform(data['Year']) 
```

```python
label_encoder_Player = LabelEncoder() 
```

```python
data['Player_id'] = label_encoder_tm.fit_transform(data['Player']) 
```

```python
data.columns
```

# Creating new variables

```python
data['PTS(t-1)'] = data.groupby('Player')['PTS'].shift(1)
data['G(t-1)'] = data.groupby('Player')['G'].shift(1)
#data['pts(t-2)'] = data.groupby('player_id')['pts'].shift(2)
data['AST(t-1)'] = data.groupby('Player')['AST'].shift(1)
#data['ast(t-2)'] = data.groupby('player_id')['ast'].shift(2)
data['TRB(t-1)'] = data.groupby('Player')['TRB'].shift(1)
#data['reb(t-2)'] = data.groupby('player_id')['reb'].shift(2)
data['FG(t-1)'] = data.groupby('Player')['FG'].shift(1)
data['FGA(t-1)'] = data.groupby('Player')['FGA'].shift(1)
data['3P(t-1)'] = data.groupby('Player')['3P'].shift(1)
data['3PA(t-1)'] = data.groupby('Player')['3PA'].shift(1)
data['2P(t-1)'] = data.groupby('Player')['2P'].shift(1)
data['2PA(t-1)'] = data.groupby('Player')['2PA'].shift(1)
data['FT(t-1)'] = data.groupby('Player')['FT'].shift(1)
data['FTA(t-1)'] = data.groupby('Player')['FTA'].shift(1)
data['DRB(t-1)'] = data.groupby('Player')['DRB'].shift(1)
data['ORB(t-1)'] = data.groupby('Player')['ORB'].shift(1)
data['TRB(t-1)'] = data.groupby('Player')['TRB'].shift(1)
data['STL(t-1)'] = data.groupby('Player')['STL'].shift(1)
data['BLK(t-1)'] = data.groupby('Player')['BLK'].shift(1)
data['TOV(t-1)'] = data.groupby('Player')['TOV'].shift(1)
data['PF(t-1)'] = data.groupby('Player')['PF'].shift(1)
data['G(t-1)'] = data.groupby('Player')['G'].shift(1)
data['MP(t-1)'] = data.groupby('Player')['MP'].shift(1)
data['Tm(t-1)'] = data.groupby('Player')['Tm'].shift(1)
```

```python
data['team_change'] = np.nan
```

```python
data['team_change'][data['Tm(t-1)'] == data['Tm']] = 0
data['team_change'][data['Tm(t-1)'] != data['Tm']] = 1
```

```python
data.corr()
```

```python
data.columns
```

```python
data.drop(columns = ['FG%', '3P%', '2P%', 'eFG%', 'FT%'], inplace = True)
```

```python
data_final = data.dropna()
```

# Predicting points

```python
#seperating independent and dependent variables
y = data_final['PTS']
X = data_final[['Player_id', 'PTS(t-1)', 'G(t-1)', 'AST(t-1)', 'TRB(t-1)', 'FG(t-1)',
       'FGA(t-1)', '3P(t-1)', '3PA(t-1)', '2P(t-1)', '2PA(t-1)', 'FT(t-1)',
       'FTA(t-1)', 'DRB(t-1)', 'ORB(t-1)', 'STL(t-1)', 'BLK(t-1)', 'TOV(t-1)',
       'PF(t-1)', 'MP(t-1)', 'Tm(t-1)', 'team_change', 'Pos', 'Age', 'Tm', 'Year']]

#importing train_test_split to create validation set
from sklearn.model_selection import train_test_split

#creating the train and validation set
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 101, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train_scaled = scale.fit_transform(X_train)
X_valid_scaled = scale.transform(X_valid)
```

# Gradient Boosting Algorithm

```python
# how to import decision tree regressor
from sklearn.ensemble import GradientBoostingRegressor

#creating the decision tree function
dt_model = GradientBoostingRegressor()
```

```python
#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid, y_valid)
```

```python
dt_model.feature_importances_
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid)
mae(y_valid, y_pred)
```

# SVM Algorithm

```python
# how to import decision tree regressor
from sklearn.svm import SVR

#creating the decision tree function
dt_model = SVR()

#fitting the model
dt_model.fit(X_train_scaled, y_train)
```

```python
#checking the training score
dt_model.score(X_train_scaled, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid_scaled, y_valid)
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid_scaled)
mae(y_valid, y_pred)
```

# AdaBoost

```python
# how to import decision tree regressor
from sklearn.ensemble import AdaBoostRegressor

#creating the decision tree function
dt_model = AdaBoostRegressor()

#fitting the model
dt_model.fit(X_train, y_train)
```
