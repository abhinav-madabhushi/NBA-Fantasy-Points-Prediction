```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
import copy
```

```python
data = pd.read_csv("Season_averages_3.csv")
del data['Unnamed: 0']
```

```python
data_season = data[['player_id', 'turnover', 'season']]
```

```python
data_season.sort_values(by = 'season', ascending = True, inplace = True)
```

# For loop to add more rows

```python
for i in range(2, 21):
    locals()["data_season_y_"+str(i)] = data_season.groupby('Player').nth(-i)
    locals()["data_season_y_"+str(i)].drop(columns = {'turnover'}, inplace = True)

    locals()["data_season_y_"+str(i)]  = pd.merge(locals()["data_season_y_"+str(i)], locals()["data_season_"+str(i+1)], on = ['Player'])

    locals()["data_season_y_"+str(i)].drop(columns = {'season_'+str(i+1)}, inplace = True)
    
    # Find the indices of the lseason non-NaN values in each row
    last_non_nan_indices = np.where(~np.isnan(locals()["data_season_final_"+str(i-1)]), np.arange(locals()["data_season_final_"+str(i-1)].shape[1])[None, :], -1).max(axis=1)
    locals()["data_season_final_"+str(i)] = copy.deepcopy(locals()["data_season_final_"+str(i-1)])

    for j in range(0, len(locals()["data_season_final_"+str(i)])):
        idx = last_non_nan_indices[j]
        locals()["data_season_final_"+str(i)].iloc[j, idx] = np.nan

    locals()["data_season_final_"+str(i)] = locals()["data_season_final_"+str(i)][locals()["data_season_final_"+str(i)]['season_1'].notnull()]
    data_season_final = data_season_final.append(locals()["data_season_final_"+str(i)], ignore_index = True)
    data_season_y = data_season_y.append(locals()["data_season_y_"+str(i)], ignore_index = True)
```
