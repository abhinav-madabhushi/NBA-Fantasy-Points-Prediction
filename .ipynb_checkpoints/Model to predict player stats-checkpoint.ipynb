```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Season_averages_3.csv")
del data["Unnamed: 0"]
```

```python
data.head()
```

```python
data['player_id'].nunique()
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns = ['last_name', 'position', 'height_feet', 'height_inches', 'weight_pounds'], inplace = True)
```

```python
data.columns
```

```python
data.corr()
```

```python
data['min'] = data['min'].str.replace(':','.').astype('float')
```

# Creating new variables

```python
data['pts(t-1)'] = data.groupby('player_id')['pts'].shift(1)
#data['pts(t-2)'] = data.groupby('player_id')['pts'].shift(2)
data['ast(t-1)'] = data.groupby('player_id')['ast'].shift(1)
#data['ast(t-2)'] = data.groupby('player_id')['ast'].shift(2)
data['reb(t-1)'] = data.groupby('player_id')['reb'].shift(1)
#data['reb(t-2)'] = data.groupby('player_id')['reb'].shift(2)
data['fgm(t-1)'] = data.groupby('player_id')['fgm'].shift(1)
data['fga(t-1)'] = data.groupby('player_id')['fga'].shift(1)
data['fg3m(t-1)'] = data.groupby('player_id')['fg3m'].shift(1)
data['fg3a(t-1)'] = data.groupby('player_id')['fg3a'].shift(1)
data['ftm(t-1)'] = data.groupby('player_id')['ftm'].shift(1)
data['fta(t-1)'] = data.groupby('player_id')['fta'].shift(1)
data['oreb(t-1)'] = data.groupby('player_id')['oreb'].shift(1)
data['dreb(t-1)'] = data.groupby('player_id')['dreb'].shift(1)
data['stl(t-1)'] = data.groupby('player_id')['stl'].shift(1)
data['blk(t-1)'] = data.groupby('player_id')['blk'].shift(1)
data['turnover(t-1)'] = data.groupby('player_id')['turnover'].shift(1)
data['pf(t-1)'] = data.groupby('player_id')['pf'].shift(1)
data['fg_pct(t-1)'] = data.groupby('player_id')['fg_pct'].shift(1)
data['fg3_pct(t-1)'] = data.groupby('player_id')['fg3_pct'].shift(1)
data['ft_pct(t-1)'] = data.groupby('player_id')['ft_pct'].shift(1)
data['games_played(t-1)'] = data.groupby('player_id')['games_played'].shift(1)
data['min(t-1)'] = data.groupby('player_id')['min'].shift(1)
```

```python
data.corr()
```

```python
data_final = data.dropna()
```

```python
data_final.columns
```

```python
data_final
```

# Prediciting points

```python
#seperating independent and dependent variables
y = data_final['pts']
X = data_final[['pts(t-1)', 'ast(t-1)', 'reb(t-1)', 'fgm(t-1)', 'fga(t-1)', 'fg3m(t-1)',
       'fg3a(t-1)', 'ftm(t-1)', 'fta(t-1)', 'oreb(t-1)', 'dreb(t-1)',
       'stl(t-1)', 'blk(t-1)', 'turnover(t-1)', 'pf(t-1)', 'fg_pct(t-1)',
       'fg3_pct(t-1)', 'ft_pct(t-1)', 'games_played(t-1)', 'min(t-1)','player_id']]

#importing train_test_split to create validation set
from sklearn.model_selection import train_test_split

#creating the train and validation set
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 101, test_size=0.25)
```

# Gradient Boosting Algorithm

```python
# how to import Gradient Boosting regressor
from sklearn.ensemble import GradientBoostingRegressor

#creating the decision tree function
dt_model = GradientBoostingRegressor()
```

```python
#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid, y_valid)
```

```python
dt_model.feature_importances_
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid)
mae(y_valid, y_pred)
```

# KNN Algorithm

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train_scaled = scale.fit_transform(X_train)
X_valid_scaled = scale.transform(X_valid)
```

```python
# how to import decision tree regressor
from sklearn.neighbors import KNeighborsRegressor

#creating the decision tree function
dt_model = KNeighborsRegressor()

#fitting the model
dt_model.fit(X_train_scaled, y_train)
```

```python
#checking the training score
dt_model.score(X_train_scaled, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid_scaled, y_valid)
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid_scaled)
mae(y_valid, y_pred)
```

# SVM Algorithm

```python
# how to import decision tree regressor
from sklearn.svm import SVR

#creating the decision tree function
dt_model = SVR()

#fitting the model
dt_model.fit(X_train_scaled, y_train)
```

```python
#checking the training score
dt_model.score(X_train_scaled, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid_scaled, y_valid)
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid_scaled)
mae(y_valid, y_pred)
```

# Building an MLP

```python
# impoting mlp regressor from sklearn
from sklearn.neural_network import MLPRegressor

# Creating instance of XGBM Regressor
clf = MLPRegressor(hidden_layer_sizes=(150,100,50),
                       max_iter = 300,activation = 'relu',
                       solver = 'adam')

# Fitting the model
clf.fit(X_train, y_train)
```

```python
#checking the training score
clf.score(X_train, y_train)
```

```python
#checking the validation score
clf.score(X_valid, y_valid)
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = clf.predict(X_valid)
mae(y_valid, y_pred)
```

# AdaBoost

```python
# how to import decision tree regressor
from sklearn.ensemble import AdaBoostRegressor

#creating the decision tree function
dt_model = AdaBoostRegressor()

#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```
