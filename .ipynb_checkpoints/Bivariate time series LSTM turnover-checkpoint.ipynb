```python
for i in range(2, 21):
    locals()["data_turnover_y_"+str(i)] = data_turnover.groupby('Player').nth(-i)
    locals()["data_turnover_y_"+str(i)].drop(columns = {'Year'}, inplace = True)

    locals()["data_turnover_y_"+str(i)]  = pd.merge(locals()["data_turnover_y_"+str(i)], locals()["data_turnover_"+str(i+1)], on = ['Player'])

    locals()["data_turnover_y_"+str(i)].drop(columns = {'turnover_'+str(i+1)}, inplace = True)
    
    # Find the indices of the lturnover non-NaN values in each row
    last_non_nan_indices = np.where(~np.isnan(locals()["data_turnover_final_"+str(i-1)]), np.arange(locals()["data_turnover_final_"+str(i-1)].shape[1])[None, :], -1).max(axis=1)
    locals()["data_turnover_final_"+str(i)] = copy.deepcopy(locals()["data_turnover_final_"+str(i-1)])

    for j in range(0, len(locals()["data_turnover_final_"+str(i)])):
        idx = last_non_nan_indices[j]
        locals()["data_turnover_final_"+str(i)].iloc[j][idx] = np.nan

    locals()["data_turnover_final_"+str(i)] = locals()["data_turnover_final_"+str(i)][locals()["data_turnover_final_"+str(i)]['turnover_1'].notnull()]
    data_turnover_final = data_turnover_final.append(locals()["data_turnover_final_"+str(i)], ignore_index = True)
    data_turnover_y = data_turnover_y.append(locals()["data_turnover_y_"+str(i)], ignore_index = True)
```

# Training LSTM Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, Dropout, ConvLSTM2D, Flatten
from tensorflow.keras.layers import Masking
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
model = Sequential()
model.add(Masking(mask_value=0., input_shape = (21, 2)))
model.add(LSTM(128, activation = 'relu', return_sequences = True))
model.add(LSTM(64, activation = 'relu'))
model.add(Dense(32))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(train_x, train_y, validation_data = (valid_x, valid_y), verbose = 2, epochs = 25)
```

```python
train_predict = model.predict(train_x)
test_predict = model.predict(test_x)
```

```python
train_predict = scalar.inverse_transform(train_predict)
test_predict = scalar.inverse_transform(test_predict)
train_y = scalar.inverse_transform(train_y)
test_y = scalar.inverse_transform(test_y)
```

```python
train_y[:,0]
```

```python
mae(train_y[:,0], train_predict[:,0])
```

```python
mae(test_y[:,0], test_predict[:,0])
```

```python
r2_score(train_y[:,0], train_predict[:,0])
```

```python
r2_score(test_y[:,0], test_predict[:,0])
```

```python
train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1, 1, 1))
```

```python
test_x = test_x.reshape((test_x.shape[0], test_x.shape[1], 1, 1, 1))
```

```python
valid_x = valid_x.reshape((test_x.shape[0], test_x.shape[1], 1, 1, 1))
```

```python
train_x.shape
```

```python
model = Sequential()
model.add(Masking(mask_value=0., input_shape=(21, 1, 1, 1)))
model.add(ConvLSTM2D(filters = 32, kernel_size = (1,1), activation = 'relu'))
model.add(Flatten())
model.add(Dense(32))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary()
```

```python
model.fit(train_x, train_y, validation_data = (test_x, test_y), verbose = 2, epochs = 20)
```

```python
train_predict = model.predict(train_x)
test_predict = model.predict(test_x)
```
