# Creating dataset

```python
data = pd.read_csv("season_averages_3.csv")
del data['Unnamed: 0']
```

```python
data
```

```python
data['Player'] = data['first_name'] + ' ' + data['last_name']
```

```python
data = data.groupby('player_id').first().reset_index()
```

```python
merged_data = pd.merge(data, data_rookie_last, on='Player')
```

```python
merged_data = pd.merge(merged_data, draft, on='Player')
```

```python
merged_data.head()
```

```python
merged_data.isnull().sum()
```

```python
merged_data['3P%'].fillna(0, inplace = True)
```

```python
del merged_data['Unnamed: 29']
```

```python
del merged_data['ORB']
del merged_data['DRB']
```

```python
merged_data['GS'].fillna(merged_data['G'], inplace = True)
merged_data['dreb'].fillna(merged_data['reb'] - merged_data['oreb'], inplace = True)
```

```python
merged_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating an instance of LabelEncoder
label_encoder_Season = LabelEncoder()

# Encoding labels in the 'Category' column
merged_data['Season'] = label_encoder_Season.fit_transform(merged_data['Season'])
```

```python
# Creating an instance of LabelEncoder
label_encoder_School = LabelEncoder()

# Encoding labels in the 'Category' column
merged_data['School'] = label_encoder_School.fit_transform(merged_data['School'])
```

```python
# Creating an instance of LabelEncoder
label_encoder_Conf = LabelEncoder()

# Encoding labels in the 'Category' column
merged_data['Conf'] = label_encoder_Conf.fit_transform(merged_data['Conf'])
```

```python
# Creating an instance of LabelEncoder
label_encoder_Class = LabelEncoder()

# Encoding labels in the 'Category' column
merged_data['Class'] = label_encoder_Class.fit_transform(merged_data['Class'])
```

```python
merged_data.columns
```

# Prediciting points

```python
#seperating independent and dependent variables
y = merged_data['pts']
X = merged_data[['Season', 'School', 'Conf', 'Class', 'G',
       'GS', 'MP', 'FG', 'FGA', 'FG%', '2P', '2PA', '2P%', '3P', '3PA', '3P%',
       'FT', 'FTA', 'FT%', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'SOS','Overall Pick']]
```

```python
#importing train_test_split to create validation set
from sklearn.model_selection import train_test_split

#creating the train and validation set
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25)
```

# Random Forest Algorithm

```python
# how to import decision tree regressor
from sklearn.ensemble import GradientBoostingRegressor

#creating the decision tree function
dt_model = GradientBoostingRegressor(max_depth = 1)

#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid, y_valid)
```

```python
dt_model.feature_importances_
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid)
mae(y_valid, y_pred)
```

# AdaBoost

```python
# how to import decision tree regressor
from sklearn.ensemble import AdaBoostRegressor

#creating the decision tree function
dt_model = AdaBoostRegressor()

#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid, y_valid)
```

```python
dt_model.feature_importances_
```

```python
from sklearn.metrics import mean_absolute_error as mae
y_pred = dt_model.predict(X_valid)
mae(y_valid, y_pred)
```
