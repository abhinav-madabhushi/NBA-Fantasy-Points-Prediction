```python
import os
import pandas as pd

# Set the directory containing your CSV files
folder_path = 'C:\\Users\\krish\\abhinav\\Machine learning Project\\Season averages prediction\\data'

# Initialize an empty list to store dataframes
dfs = []

# Loop through all files in the directory
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Read each CSV file into a dataframe and append to the list
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate all dataframes in the list along the rows
result = pd.concat(dfs, ignore_index=True)

# Now 'result' contains the concatenated dataframe
```

```python
del result["Unnamed: 0"]
```

```python
result.drop(columns = {'team_id', 'team_abbreviation', 'team_city',
       'team_conference', 'team_division', 'team_full_name', 'team_name'}, inplace = True)
```

```python
result.to_csv("Season_averages_3.csv")
```
