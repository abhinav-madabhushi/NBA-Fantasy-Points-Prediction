```python
import pandas as pd

# The columns you want to weight‑average
stat_cols = [
    'min', 'fgm', 'fga', 'fg3m', 'fg3a', 'ftm', 'fta',
    'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk',
    'turnover', 'pf', 'pts', 'fg_pct', 'fg3_pct', 'ft_pct'
]

def weighted_avg(group):
    # total games in this player×season block
    total_gp = group['games_played'].sum()
    # build a dict of weighted means
    return pd.Series({
        stat: (group[stat] * group['games_played']).sum() / total_gp
        for stat in stat_cols
    })

# apply it:
data_2 = (
    data
      .groupby(['player_id','season'])
      .apply(weighted_avg)
      .reset_index()
)
```

```python
import pandas as pd
import re

df = team.copy()
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Make a sortable start-year from Season (handles "2024-25" or 2025)
def season_start(s):
    if pd.isna(s): 
        return pd.NA
    m = re.search(r'(\d{4})', str(s))
    return int(m.group(1)) if m else pd.NA

df['season_start'] = df['Season'].apply(season_start)

latest = df['season_start'].max()
prev   = latest - 1

# Age map from previous season (by player name)
prev_age_map = (df[(df['season_start'] == prev) & df['Age'].notna()]
                .set_index('full_name')['Age'])

# Fill for rows in the latest season where Age is missing
mask = (df['season_start'] == latest) & (df['Age'].isna())
df.loc[mask, 'Age'] = df.loc[mask, 'full_name'].map(prev_age_map).add(1)

# If still missing (no prior season), default to 20
df.loc[mask & df['Age'].isna(), 'Age'] = 20

# Optional: make Age a nullable integer and drop helper column
df['Age'] = df['Age'].astype('Int64')
df = df.drop(columns=['season_start'])
```

```python
year_list = [2025]
X_train = X[~X['season'].isin(year_list)]
y_train = y[~y['season'].isin(year_list)]
X_train.set_index('season', inplace=True)
del y_train['season']
X_valid = X[X['season'].isin(year_list)]
y_valid = y[y['season'].isin(year_list)]
X_valid.set_index('season', inplace=True)
del y_valid['season']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.multioutput import MultiOutputRegressor

# Wrap the base regressor
model = MultiOutputRegressor(GradientBoostingRegressor())

# Fit the model
model.fit(X_train, y_train)
```
