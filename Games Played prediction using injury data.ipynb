```python
data_games_played_1 = data_games_played.groupby(['player_id', 'season'])['games_played'].sum().reset_index()
```

```python
import pandas as pd

# The columns you want to weight‑average
stat_cols = [
    'min', 'fgm', 'fga', 'fg3m', 'fg3a', 'ftm', 'fta',
    'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk',
    'turnover', 'pf', 'pts', 'fg_pct', 'fg3_pct', 'ft_pct'
]

def weighted_avg(group):
    # total games in this player×season block
    total_gp = group['games_played'].sum()
    # build a dict of weighted means
    return pd.Series({
        stat: (group[stat] * group['games_played']).sum() / total_gp
        for stat in stat_cols
    })

# apply it:
data_games_played_2 = (
    data_games_played
      .groupby(['player_id','season'])
      .apply(weighted_avg)
      .reset_index()
)
```

```python
data_games_played = data_games_played_1.merge(data_games_played_2, on = ['player_id', 'season'])
```

```python
start_year = 1960
# Function to generate seasons based on the occurrence of '1' in 'Rk'
def generate_seasons(row, season_marker):
    global start_year
    if row['Rk'] == season_marker:
        start_year += 1
    return start_year
```
