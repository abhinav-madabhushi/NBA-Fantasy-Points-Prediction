```python
start_year = 1960
# Function to generate seasons based on the occurrence of '1' in 'Rk'
def generate_seasons(row, season_marker):
    global start_year
    if row['Rk'] == season_marker:
        start_year += 1
    return start_year
```

```python
# Apply function to create the new 'Season' column
team['Season'] = team.apply(generate_seasons, season_marker=1, axis=1)
```
