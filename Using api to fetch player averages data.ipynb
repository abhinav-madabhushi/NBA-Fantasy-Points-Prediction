```python
if response.status_code == 200:
    for i in range(1957, 1960):
        # Creating a DataFrame
        df = pd.DataFrame()
        df_missing = pd.DataFrame({'player_id':[], 'season':[]})
        for j in x:
            print(i,j)
            response = requests.get('https://api.balldontlie.io/v1/season_averages?season={}&player_ids[]={}'.format(i,j), headers=headers)
            try:
                response_json = response.json()
            except JSONDecodeError:
                print('Error decoding JSON for player ID', j, 'in season', i)
                new_row = pd.Series({'player_id':j, 'season':i})
                df_missing = df_missing.append(new_row, ignore_index=True)
                time.sleep(60)
                response = requests.get('https://api.balldontlie.io/v1/season_averages?season={}&player_ids[]={}'.format(i,j))
            if (len((pd.DataFrame(response.json()['data']))) != 0):
                temp_df_1 = (pd.DataFrame(response.json()['data']))[["games_played", "player_id", "season", "min", "fgm", "fga", "fg3m", "fg3a", "ftm", "fta", "oreb", "dreb", "reb", "ast", "stl", "blk", "turnover", "pf", "pts", "fg_pct", "fg3_pct", "ft_pct"]]
                df = df.append(temp_df_1, ignore_index=True)
        df_1 = df.merge(player_info, on = ['player_id'])
        df_1.to_csv("Season_averages_" + str(i) + ".csv")
else:
    print('Error', response.status_code)
```
