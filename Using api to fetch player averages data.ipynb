```python
import pandas as pd
import requests
import time

df = pd.DataFrame()
for i in range(2023, 2024):
    for j in x:
        print(i, j)
        url = f'https://api.balldontlie.io/v1/stats?seasons[]={i}&player_ids[]={j}&per_page=100'
        response = requests.get(url, headers=headers)
        try:
            response_json = response.json()
            data = response_json['data']
        except Exception as e:
            print('Error decoding JSON for player ID', j, 'in season', i)
            continue

        for game_stat in data:
            flat_record = {
                'game_id': game_stat['game']['id'],
                'player_id': game_stat['player']['id'],
                'season': game_stat['game']['season'],
                'pts': game_stat['pts'],
                'ast': game_stat['ast'],
                'reb': game_stat['reb'],
                'fgm': game_stat['fgm'],
                'fga': game_stat['fga'],
                'fg3m': game_stat['fg3m'],
                'fg3a': game_stat['fg3a'],
                'ftm': game_stat['ftm'],
                'fta': game_stat['fta'],
                'stl': game_stat['stl'],
                'blk': game_stat['blk'],
                'turnover': game_stat['turnover'],
                'pf': game_stat['pf'],
                'min': game_stat['min'],
                'fg_pct': game_stat['fg_pct'],
                'fg3_pct': game_stat['fg3_pct'],
                'ft_pct': game_stat['ft_pct'],
                'date': game_stat['game']['date']
            }
            df = pd.concat([df, pd.DataFrame([flat_record])], ignore_index=True)
```

```python
df
```

```python
df['min'] = df['min'].astype('int')
```

```python
df = df[df['min'] != 0]
```

```python
agg_columns = [
    'pts', 'ast', 'reb', 'fgm', 'fga', 'fg3m', 'fg3a', 'ftm', 'fta',
    'stl', 'blk', 'turnover', 'pf', 'min', 'fg_pct', 'fg3_pct', 'ft_pct'
]

# Perform groupby and aggregate
season_stats = df.groupby(['player_id', 'season'])[agg_columns].mean().reset_index()

# Optional: Add number of games played
season_stats['games_played'] = df.groupby(['player_id', 'season']).size().values
```

```python
season_stats = season_stats[season_stats['min'] != 0]
```

```python
season_stats.to_csv("Data/Season_averages_2023.csv")
```
