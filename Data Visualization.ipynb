## Still injured versus gamed played 

```python
import matplotlib.pyplot as plt

# 1) Aggregate total games played by injury status
agg = data.groupby('still_injured')['games_played'].mean()

ucla_light_blue = "#53A4DE"
ucla_gray = "#666666"

palette = [ucla_light_blue, ucla_gray]

# 2) Plot bar chart
plt.figure(figsize=(6, 4))
plt.bar(agg.index.astype(str), agg.values, color = palette)
plt.xlabel("Still Injured")
plt.ylabel("Total Games Played")
plt.title("Games Played by Injury Status")
plt.show()
```

# Graphs for rookie data

```python
data_rookie = data[data['Rookie'] == 1]
```

```python
data_rookie_1 = data_rookie[(data_rookie['season'] != 2025) & (data_rookie['season'] != 2024)]
```

```python
import matplotlib.pyplot as plt

# List of stats to compare
stats_pred = [
    'games_played', 'pts', 'fgm', 'fga', 'fg3m',
    'fg3a', 'ftm', 'fta', 'reb', 'ast',
    'stl', 'blk', 'turnover'
]

# 1) Compute correlation for each stat between (t‑1) and current
corrs = {}
for stat in stats_pred:
    lag_col = f"{stat}(t-1)"
    if lag_col in data_rookie_1.columns and stat in data_rookie_1.columns:
        corrs[stat] = data_rookie_1[lag_col].corr(data_rookie_1[stat])

# Convert to a pandas Series and sort by value
import pandas as pd
corr_series = pd.Series(corrs).sort_values()

# 2) Plot as a horizontal bar chart
plt.figure(figsize=(8, 6))
plt.barh(corr_series.index, corr_series.values, color='teal')
plt.xlabel("Pearson R between statistic(t‑1) and statistic")
plt.title("Lagged vs. Current Season Correlation for Rookies")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Make draft‐position bins. Adjust cut points as needed.
#    Here “1–10”, “11–20”, “21–30”, and “31+”
bins = [0, 10, 20, 30, data_rookie['Overall Pick'].max()]
labels = ["1–10", "11–20", "21–30", "31+"]
data_rookie['pick_group'] = pd.cut(
    data_rookie['Overall Pick'], 
    bins=bins, 
    labels=labels
)

stats_pred = [
    'games_played', 'pts', 'fgm', 'fga', 'fg3m',
    'fg3a', 'ftm', 'fta', 'reb', 'ast',
    'stl', 'blk', 'turnover'
]

# 2) For each stat, draw a boxplot of “pick_group” vs that stat
plt.figure(figsize=(12, 3 * len(stats_pred)))
for i, stat in enumerate(stats_pred, 1):
    ax = plt.subplot(len(stats_pred), 1, i)
    sns.boxplot(
        x='pick_group',
        y=stat,
        data=data_rookie,
        palette="pastel",
        ax=ax
    )
    ax.set_xlabel("")           # remove repeated x‐label
    ax.set_ylabel(stat)
    ax.set_title(f"{stat} by Draft Pick Group")
plt.tight_layout()
plt.show()
```
