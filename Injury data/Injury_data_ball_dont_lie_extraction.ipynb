```python
import requests
import pandas as pd
import time

API_KEY = "YOUR_API_KEY"

# Base URL
base_url = "https://api.balldontlie.io/v1/player_injuries"
headers = {
    "Authorization": f"a06fc0a0-aa21-40a5-8abf-476308868dc9"
}

# Parameters
per_page = 100  # max allowed
page = 1
all_data = []

while True:
    response = requests.get(base_url, params={"page": page, "per_page": per_page}, headers=headers)
    if response.status_code != 200:
        print(f"Failed to fetch page {page}. Status code: {response.status_code}")
        break

    data = response.json()
    injuries = data.get("data", [])

    if not injuries:
        break  # no more data

    all_data.extend(injuries)
    print(f"Fetched page {page} with {len(injuries)} records.")
    page += 1
    time.sleep(0.25)  # avoid overloading the API

# Convert to DataFrame
injuries_df = pd.json_normalize(all_data)

# Display the top rows
print(injuries_df.head())
```
